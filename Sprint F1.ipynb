```python
#Sprint
from bs4 import BeautifulSoup
import requests
import pandas as pd
urls=['https://chicanef1.com./race.pl?year=2022&gp=Austrian%20GP&r=1&type=sprint', \
    'https://chicanef1.com./race.pl?year=2022&gp=Emilia%20Romagna%20GP&r=1&type=sprint',
    'https://chicanef1.com./race.pl?year=2022&gp=S%E3o%20Paulo%20GP&r=1&type=sprint']
index=1
for url in urls:
        page=requests.get(url)
        soup= BeautifulSoup(page.text, 'html.parser')
        table=soup.find('td',class_='yellowborder')
        table_columns= table.find_all('th')[:8]
        table_titles=[title.text.strip() for title in table_columns]

        df= pd.DataFrame(columns=table_titles)
        column_data=table.find_all('tr')

        for row in column_data[1:]:
            row_data = row.find_all('td')
            individual_row_data = [data.text.strip() for data in row_data]
            print(individual_row_data)
            length=len(df)
            df.loc[length]= individual_row_data
        df.to_csv(f'/Users/valent/Documents/Python/sprint_scrape{index}.csv',index= False)
        index+=1
```
